```
from collections import deque

clientes = ["Ana", "Bruno", "Carlos", "Daniel", "Elena", "Fernanda", "Gabriel", "Helena"]

fila = deque()

for c in clientes:
    fila.append(c)

print("Fila inicial:", fila)

for _ in range(3):
    atendido = fila.popleft()
    print("Atendido:", atendido)

fila.append("Igor")

print("Fila atual:", fila)
```
