```
def primo(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

primos = [x for x in range(2, 101) if primo(x)]
print(primos)
```

```
estoque = [
    {"produto": "Caneta", "quantidade": 5},
    {"produto": "Caderno", "quantidade": 12},
    {"produto": "Borracha", "quantidade": 3},
    {"produto": "Mochila", "quantidade": 20},
]

baixo_estoque = [p for p in estoque if p["quantidade"] < 10]

print("Produtos com estoque baixo:")
for item in baixo_estoque:
    print(f"{item['produto']} — {item['quantidade']}")
```

```
historico = [
    "google.com",
    "youtube.com",
    "github.com",
    "google.com",
    "stackoverflow.com",
    "github.com"
]

vistos = set()
resultado = []

for site in reversed(historico):
    if site not in vistos:
        resultado.append(site)
        vistos.add(site)

resultado = list(reversed(resultado))

print("Histórico sem duplicatas:", resultado)
```

```
notas = [6.5, 8.0, 7.2, 5.0, 9.1, 4.8, 7.0]

media = sum(notas) / len(notas)
maior = max(notas)
aprovados = len([n for n in notas if n >= 7])

print("Média:", media)
print("Maior nota:", maior)
print("Aprovados:", aprovados)
```

```
palavras = ["python", "código", "python", "dados", "código", "python"]

frequencia = {}

for p in palavras:
    p = p.lower()
    frequencia[p] = frequencia.get(p, 0) + 1

print("Frequência de palavras:")
for palavra, qtd in frequencia.items():
    print(f"{palavra}: {qtd}")
```
