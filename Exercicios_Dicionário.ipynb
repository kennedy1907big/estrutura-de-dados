```
quadrados = {}

for i in range(1, 6):
    quadrados[i] = i**2

print(quadrados)
```

```
d = {1: 10, 2: 20, 3: 30, 4: 40}

print(sum(d.values()))
```

```
notas = {"Ana": 8, "João": 9, "Maria": 7}

maior = max(notas, key=notas.get)

print(maior)
```

```
chaves = ["a", "b", "c"]
valores = [1, 2, 3]

dicionario = dict(zip(chaves, valores))

print(dicionario)
```
