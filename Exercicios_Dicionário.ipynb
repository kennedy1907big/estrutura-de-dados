```
import unicodedata

def normalizar(texto):
    return ''.join(c for c in unicodedata.normalize('NFD', texto)
                   if unicodedata.category(c) != 'Mn').lower()

def contar(texto):
    texto = normalizar(texto)
    d = {}
    for c in texto:
        if c.isalpha():
            d[c] = d.get(c, 0) + 1
    return d

print(contar("Programação"))
```

```
alunos = [
    {"nome": "Ana", "idade": 20, "nota": 8},
    {"nome": "Bruno", "idade": 22, "nota": 6},
    {"nome": "Carlos", "idade": 19, "nota": 9}
]

media = sum(a["nota"] for a in alunos) / len(alunos)

maior = max(alunos, key=lambda a: a["nota"])

menor = min(alunos, key=lambda a: a["nota"])

print("Média:", media)
print("Maior nota:", maior)
print("Menor nota:", menor)
```

```
estoque = {
    "banana": 10,
    "maçã": 5
}
venda = {"banana": 3, "maçã": 2}

for produto, qtd in venda.items():
    if produto in estoque:
        estoque[produto] -= qtd

print(estoque)
```
