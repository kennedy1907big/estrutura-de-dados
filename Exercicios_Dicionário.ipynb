```
nomes = ["Ana", "Bruno", "Ana", "Carlos", "Bruno"]

freq = {}

for nome in nomes:
    freq[nome] = freq.get(nome, 0) + 1

print(freq)
```

```
def contar_caracteres(texto):
    freq = {}

    for c in texto.lower():
        freq[c] = freq.get(c, 0) + 1

    return freq

print(contar_caracteres("Python"))
```

```
d = {"a": 1, "b": 2, "c": 1}

invertido = {}

for k, v in d.items():
    invertido.setdefault(v, []).append(k)

print(invertido)
```

```
pessoa = {
    "nome": "Ana",
    "endereco": {
        "rua": "A",
        "cidade": "Três Lagoas"
    },
    "contatos": {
        "telefone": "9999",
        "email": "ana@email.com"
    }
}

print(pessoa)
```

```
cidade = pessoa.get("endereco", {}).get("cidade")

print(cidade)
```

```
def imprimir(d, nivel=0):
    for k, v in d.items():
        print("  " * nivel + str(k))
        if isinstance(v, dict):
            imprimir(v, nivel + 1)
        else:
            print("  " * (nivel + 1) + str(v))

imprimir(pessoa)
```
