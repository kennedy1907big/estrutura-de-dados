```
pessoas = [
    {"nome": "Carlos", "idade": 20},
    {"nome": "Ana", "idade": 25},
    {"nome": "Marcos", "idade": 30}
]

for pessoa in pessoas:
    print(pessoa["nome"])
```

```
pessoas = [
    {"nome": "Carlos", "idade": 20},
    {"nome": "Ana", "idade": 25},
    {"nome": "Marcos", "idade": 30}
]

soma = 0

for pessoa in pessoas:
    soma += pessoa["idade"]

media = soma / len(pessoas)

print("Média:", media)
```

```
pessoas = [
    {"nome": "Carlos", "idade": 20},
    {"nome": "Ana", "idade": 25}
]

nova_pessoa = {"nome": "Julia", "idade": 22}

pessoas.append(nova_pessoa)

print(pessoas)
```

```
pessoas = [
    {"nome": "Carlos", "idade": 20},
    {"nome": "Ana", "idade": 25},
    {"nome": "Julia", "idade": 22}
]

for pessoa in pessoas:
    if pessoa["nome"] == "Ana":
        pessoas.remove(pessoa)

print(pessoas)
```

```
import json

pessoas = [
    {"nome": "Carlos", "idade": 20},
    {"nome": "Ana", "idade": 25}
]

json_string = json.dumps(pessoas, indent=4)

print(json_string)
```

```
import json

pessoas = [
    {"nome": "Carlos", "idade": 20},
    {"nome": "Ana", "idade": 25}
]

with open("pessoas.json", "w", encoding="utf-8") as arquivo:
    json.dump(pessoas, arquivo, indent=4)

print("Arquivo salvo com sucesso.")
```

```
import json

with open("pessoas.json", "r", encoding="utf-8") as arquivo:
    dados = json.load(arquivo)

print(dados)
```

```
pessoas = [
    {"nome": "Carlos", "idade": 20},
    {"nome": "Ana", "idade": 25}
]

for pessoa in pessoas:
    if pessoa["nome"] == "Carlos":
        print(pessoa)
```

```
pessoas = [
    {"nome": "Carlos", "idade": 20},
    {"nome": "Ana", "idade": 25}
]

for pessoa in pessoas:
    if pessoa["nome"] == "Carlos":
        pessoa["idade"] = 21

print(pessoas)
```
