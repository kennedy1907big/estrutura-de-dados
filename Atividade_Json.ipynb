```
import json
```

```
dados = {
    "nome": "Kennedy",
    "idade": 25,
    "cidade": "Tres Lagoas"
}

json_str = json.dumps(dados)
print(json_str)
```

```
print(json.dumps(dados, indent=4))
```

```
json_texto = '{"nome": "Kennedy", "idade": 25}'
dicionario = json.loads(json_texto)

print(dicionario)
print(type(dicionario))
```

```
json_manual = '{"produto": "Coca", "preco": 5.50}'
dados = json.loads(json_manual)

print(dados)
```

```
with open("dados.json", "w") as arquivo:
    json.dump(dados, arquivo)
```

```
with open("dados.json", "r") as arquivo:
    conteudo = json.load(arquivo)

print(conteudo)
```

```
numeros = [1, 2, 3, 4, 5]
json_lista = json.dumps(numeros)

print(json_lista)
```

```
dados = json.loads(json_lista)

print(type(dados))
```

```
json_exemplo = '{"ativo": true, "erro": false, "valor": null}'
dados = json.loads(json_exemplo)

print(dados)
```

```
dados = {
    "numeros": {1, 2, 3}  # set
}

json.dumps(dados)
```

o valor {1, 2, 3} é um conjunto (set) e o JSON não tem represnetação equivalente para isso, ou seja, antes tem tem que converter o conjunto para ista antes de serializar, abaixo esta a forma corrigida se precisar

```
dados = {
    "numeros": list({1, 2, 3})
}
json.dumps(dados)
```

```
aluno = {
    "nome": "Ana",
    "nota": 8.5,
    "aprovado": True
}

print(json.dumps(aluno, indent=4))
```

```
print(json.dumps(aluno, indent=4, ensure_ascii=False))
```

```
nomes = ["Ana", "João", "Carlos"]
json_nomes = json.dumps(nomes)

dados = json.loads(json_nomes)

for nome in dados:
    print(nome)
```

```
json_num = '[10, 20, 30]'
dados = json.loads(json_num)

print(sum(dados))
```

```
import json

# Criando o arquivo
dados = {
    "nome": "Teste",
    "idade": 30
}

with open("teste.json", "w") as f:
    json.dump(dados, f)

# Leitura
with open("teste.json") as f:
    print(json.load(f))
```

```
dados = {
    "nome": "Carlos",
    "idade": 40,
    "cidade": "SP"
}

for chave, valor in dados.items():
    print(chave, ":", valor)
```

```
dados = {
    "nomes": ["Ana", "João"]
}

print(json.dumps(dados))
```

```
json_obj = '''
[
    {"nome": "Ana"},
    {"nome": "João"}
]
'''

dados = json.loads(json_obj)

for pessoa in dados:
    print(pessoa["nome"])
```

```
banco = [
    {"id": 1, "nome": "Kennedy", "saldo": 1000},
    {"id": 2, "nome": "Ana", "saldo": 1500}
]

with open("banco.json", "w") as f:
    json.dump(banco, f, indent=4)

print("Banco criado!")
```
