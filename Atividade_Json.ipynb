```
alunos = [
    {"nome": "Maria", "nota": 9.0},
    {"nome": "João", "nota": 7.5},
    {"nome": "Ana", "nota": 5.0}
]

for aluno in alunos:
    print(aluno["nome"], "-", aluno["nota"])
```

```
for aluno in alunos:
    if aluno["nota"] >= 7:
        print(aluno)
```

```
nome_busca = "Ana"

for aluno in alunos:
    if aluno["nome"] == nome_busca:
        print(aluno)
```

```
for aluno in alunos:
    if aluno["nome"] == "João":
        aluno["nota"] = 7
```

```
alunos.append({"nome": "Lucas", "nota": 10})
```

```
nome_remover = "Ana"

alunos = [a for a in alunos if a["nome"] != nome_remover]
```

```
media = sum(a["nota"] for a in alunos) / len(alunos)
print(media)
```

```
maior = max(alunos, key=lambda a: a["nota"])
print(maior)
```

```
ordenados = sorted(alunos, key=lambda a: a["nota"])
print(ordenados)
```

```
import json
from collections import Counter

usuarios = [
    {
        "nome": "Ana",
        "contatos": {
            "email": "ana@email.com",
            "telefone": "1111"
        }
    },
    {
        "nome": "João",
        "contatos": {
            "email": "joao@email.com",
            "telefone": "2222"
        }
    }
]
```

```
print(usuarios[0]["contatos"]["email"])
```

```
flatten = []

for u in usuarios:
    novo = {
        "nome": u["nome"],
        "email": u["contatos"]["email"],
        "telefone": u["contatos"]["telefone"]
    }
    flatten.append(novo)

print(flatten)
```

```
alunos = [
    {"nome": "Ana", "turma": "A"},
    {"nome": "João", "turma": "B"},
    {"nome": "Carlos", "turma": "A"}
]

grupo = {}

for a in alunos:
    turma = a["turma"]
    grupo.setdefault(turma, []).append(a)

print(grupo)
```

```
nomes = [a["nome"] for a in alunos]

contagem = Counter(nomes)
print(contagem)
```

```
vendas = [
    {"vendedor": "Ana", "valor": 100},
    {"vendedor": "Ana", "valor": 200},
    {"vendedor": "João", "valor": 150}
]

total = {}

for v in vendas:
    total[v["vendedor"]] = total.get(v["vendedor"], 0) + v["valor"]

print(total)
```
